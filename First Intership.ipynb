```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = r"C:\Users\This PC\Documents\Batch34_List_Final_Formatted.xlsx"

 
df = pd.read_excel(file_path)


df.columns = df.columns.str.strip()


domain_counts_by_year = df.groupby(['Domain', 'Academic Year']).size().reset_index(name="Student Count")


plt.figure(figsize=(14, 7))


sns.set_theme(style="whitegrid")
bars = sns.barplot(
    x="Domain", 
    y="Student Count", 
    hue="Academic Year", 
    data=domain_counts_by_year, 
    palette="Set2"
)


plt.title("Number of Students per Domain by Academic Year", fontsize=16, fontweight="bold")
plt.xlabel("Domain", fontsize=14)
plt.ylabel("Number of Students", fontsize=14)


plt.xticks(rotation=45, ha="right", fontsize=12)


plt.grid(axis="y", linestyle="--", alpha=0.7)


for bar in bars.patches:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2, 
        yval + 0.05, 
        round(yval, 2), 
        ha="center", 
        va="bottom", 
        fontsize=12
    )


plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


file_path = r"C:\Users\This PC\Documents\Batch34_List_Final_Formatted.xlsx"
df = pd.read_excel(file_path)


df.columns = df.columns.str.strip()


domain_col = "Domain"


domain_counts = df[domain_col].value_counts()


domain_counts = domain_counts.sort_values(ascending=False)


sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))


ax = sns.barplot(
    x=domain_counts.index,
    y=domain_counts.values,
    hue=domain_counts.index,
    palette="Blues_d",
    legend=False
)


plt.xticks(rotation=45, ha='right')


for i, value in enumerate(domain_counts.values):
    plt.text(i, value + 5, str(value), ha='center', va='bottom', fontweight='bold')


plt.title("Most Popular Internship Domains", fontsize=16, fontweight='bold')
plt.xlabel("Internship Domain", fontsize=12)
plt.ylabel("Number of Students", fontsize=12)

plt.tight_layout()
plt.show()
```
